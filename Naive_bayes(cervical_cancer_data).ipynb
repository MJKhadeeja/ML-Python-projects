```python
x=cc_data.drop(columns=['ca_cervix'],axis=1)
y=cc_data['ca_cervix']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
x_train_predict=model.predict(x_train)
train_data_accuracy=accuracy_score(y_train,x_train_predict)
```

```python
print('the accuracy of train data is',train_data_accuracy)
```

```python
x_test_predict=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_predict)
```

```python
print('test data accuracy is',test_data_accuracy)
```
