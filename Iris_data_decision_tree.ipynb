```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
x=iris_data.drop('Species',axis=1)
y=iris_data['Species']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
clf=DecisionTreeClassifier(random_state=42)  #create a decision tree classifier
```

```python
clf.fit(x_train,y_train)    #train the classifier on the train data set
```

```python
y_pred = clf.predict(x_test)  #make prediction on test data
```

```python
y_pred1=clf.predict(x_train)
```

```python
#calculate the accuracy of the classifier
accuracy=accuracy_score(y_test,y_pred)

print(accuracy)
```

```python
accuarcy=accuracy_score(y_train,y_pred1)
print(accuarcy)
```

```python
#display classification report
print(classification_report(y_test,y_pred))
```

```python
input_data=(94,5.0,2.3,3.3,1.0)
input_data_array=np.array(input_data)
input_data_reshaped=input_data_array.reshape(1,-1)

prediction=clf.predict(input_data_reshaped)
print(prediction[0])

if(prediction[0]==0):
  print('sample is iris setosa')
elif(prediction[0]==1):
  print('sample is iris versicolor')
else:
  print('sample is iris verginica')
```
