```python
x=wine_data.drop('quality',axis=1)
print(x)
```

```python
y=wine_data['quality'].apply(lambda y_value:1 if y_value>= 7 else 0)  #when the y value greter or equal to 7 the quality of the wine is good otherwise bad
```

```python
print(y)
```

```python
wine_data.value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print('Accuracy on training data is ',training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print('accuracy on test data is',test_data_accuracy)
```

```python
x_test.to_csv('wine_data.csv',index=False)
#from google.colab import files
#files.download('wine_data.csv')
```
